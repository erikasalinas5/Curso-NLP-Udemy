```python
import pandas as pd
import numpy as np
import nltk
```

```python
datos = pd.read_csv('peliculas.tsv',sep='\t')
```

```python
datos.head(5)
datos.dropna(inplace=True)
```

```python
blancos = []
for i, label,revision in datos.itertuples():
    if revision.isspace():
        blancos.append(i)
```

```python
datos.drop(blancos,inplace=True)
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer as sia
```

```python
analizador = sia()
datos['analisis']=datos['revision'].apply(lambda revision: analizador.polarity_scores(revision))
```

```python
datos['analisis'][1]
```

```python
datos['resultado']=datos['analisis'].apply(lambda d: d['compound'])
```

```python
datos.head(5)
```

```python
datos['prediccion'] = datos['resultado'].apply(lambda resultado:'pos' if resultado>=0 else 'neg')
```

```python
datos.head(5)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
accuracy_score(datos['etiqueta'],datos['prediccion'])
```

```python
print(classification_report(datos['etiqueta'],datos['prediccion']))
```

```python
print(confusion_matrix(datos['etiqueta'],datos['prediccion']))
```
